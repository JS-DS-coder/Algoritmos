```python
import numpy as np
```

```python

class Perceptron_Multiple:
    def __init__(self, entrada1, objetivo):
        self.learning_rate = 0.2
        self.entrada = entrada1
        self.objetivo = objetivo
        self.W1 = np.random.uniform(size=(2, 2))
        self.W2 = np.random.uniform(size=(2, 1))
        self.bias1 = np.random.uniform(size=(1,2))
        self.bias2 = np.random.uniform(size=(1,1))
        self.perdidas = []
    

    def backwardPropagation(self,salida,salidacapa1):
        
        DerivadaSigmoid = self.derivadasigmoid(salida)

        perdida = 0.5 * (self.objetivo - salida) ** 2
        self.perdidas.append(np.sum(perdida))

        error = (self.objetivo - salida)

     
        gradiente1 = self.entrada.T @ (((error * DerivadaSigmoid) * self.W2.T) * self.derivadasigmoid(salidacapa1))

       
        gradiente2 = salida.T @ (error * DerivadaSigmoid)

  
        self.W1 = self.W1 + (self.learning_rate * gradiente1)
        self.W2 = self.W2 + (self.learning_rate * gradiente2)

        self.bias1 = self.bias1 + np.sum(self.learning_rate * ((error * DerivadaSigmoid) * self.W2.T) * self.derivadasigmoid(salidacapa1), axis=0)
        self.bias2 = self.bias2 + np.sum(self.learning_rate * error * DerivadaSigmoid, axis=0)

    def sigmoid(self, x):
        #Funcion sigmoid
        return 1 / (1 + np.exp(-x))

    def derivadasigmoid(self, x):
      #La primera derivada de la funcion Sigmoid
        return x * (1 - x)

    def forwardPropagation(self, x):
      
        self.Capaoculta1 = np.dot(x, self.W1) + self.bias1
        self.Capaoculta1_out = self.sigmoid(self.Capaoculta1)

        self.Capaoculta2 = np.dot(self.Capaoculta1_out, self.W2) + self.bias2
        self.Capaoculta2_out = self.sigmoid(self.Capaoculta2)

        self.backwardPropagation(self.Capaoculta2_out,self.Capaoculta1_out)

    
    def entrenar(self,X,K,print):
        for i in range(K):
          self.forwardPropagation(self.entrada)
         # if i % 500 == 0:
         #   print(i)
            #print("Error: ", self.perdidas[i])
        return self.perdidas
```
